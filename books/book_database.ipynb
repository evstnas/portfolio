# Анализ базы данных сервиса для чтения книг по подписке

Компания купила крупный сервис для чтения книг по подписке. 

Задача — проанализировать базу данных. В ней — информация о книгах, издательствах, авторах, а также пользовательские обзоры книг. Эти данные помогут сформулировать ценностное предложение для нового продукта.

### Описание данных

**Таблица `books`**

Содержит данные о книгах:

- `book_id` — идентификатор книги;
- `author_id` — идентификатор автора;
- `title` — название книги;
- `num_pages` — количество страниц;
- `publication_date` — дата публикации книги;
- `publisher_id` — идентификатор издателя.

**Таблица `authors`**

Содержит данные об авторах:

- `author_id` — идентификатор автора;
- `author` — имя автора.

**Таблица `publishers`**

Содержит данные об издательствах:

- `publisher_id` — идентификатор издательства;
- `publisher` — название издательства;

**Таблица `ratings`**

Содержит данные о пользовательских оценках книг:

- `rating_id` — идентификатор оценки;
- `book_id` — идентификатор книги;
- `username` — имя пользователя, оставившего оценку;
- `rating` — оценка книги.

**Таблица `reviews`**

Содержит данные о пользовательских обзорах на книги:

- `review_id` — идентификатор обзора;
- `book_id` — идентификатор книги;
- `username` — имя пользователя, написавшего обзор;
- `text` — текст обзора.

## Чтение файла

```python
import pandas as pd
from sqlalchemy import create_engine 
```

- Посчитайте, сколько книг вышло после 1 января 2000 года;

```python
sql('''
    SELECT COUNT(*) AS books_count
    FROM books
    WHERE publication_date > '2000-01-01'
''')
```

В базе есть 819 книг, которые были опубликованы с 2000 года, при этом всего - 1000 книг.

- Для каждой книги посчитайте количество обзоров и среднюю оценку;

```python
sql('''
    SELECT b.book_id, 
           b.title,
           br.review_count,
           rb.avg_rating
    FROM books AS b
    LEFT JOIN (
        SELECT rev.book_id,
               COUNT(rev.review_id) AS review_count
        FROM reviews AS rev
        GROUP BY rev.book_id
    ) AS br ON b.book_id = br.book_id
    LEFT JOIN ( 
        SELECT rat.book_id,
               AVG(rat.rating) AS avg_rating
        FROM ratings as rat
        GROUP BY rat.book_id
    ) AS rb ON b.book_id = rb.book_id
    ORDER BY book_id
''')
```

Посчитали кол-во обзоров и среднюю оценку для каждой книги, есть те, у которых нет оценок и отзывов.

- Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры;

```python
sql('''
    SELECT p.publisher_id,
           p.publisher,
           COUNT(b.book_id) AS books_count
    FROM publishers as p
    JOIN books AS b ON b.publisher_id = p.publisher_id
    WHERE b.num_pages > 50 
    GROUP BY p.publisher_id
    ORDER BY books_count DESC
    limit 1
''')
```

Издательство, которое выпустило наибольшее число книг (42 шт. из базы) толще 50 стр. - Penguin Books.

- Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками;

```python
sql('''
    SELECT a.author_id, a.author, AVG(b.agv_book_rating) as avg_rating
    FROM authors as a
    JOIN (
        SELECT b.book_id, b.author_id, COUNT(*) as ratings_count, AVG(r.rating) as agv_book_rating
        FROM ratings as r
        JOIN books as b ON r.book_id = b.book_id
        GROUP BY b.book_id, b.author_id
        ORDER BY b.book_id
    ) as b ON a.author_id = b.author_id
    WHERE ratings_count >= 50
    GROUP BY a.author_id
    ORDER BY avg_rating DESC
    LIMIT 1
''')
```

В этой базе самым популярным автором является Джоан Роулинг.

- Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
sql('''
    SELECT AVG(r.review_count) as avg_rating
    FROM (
        SELECT COUNT(rev.review_id) AS review_count
        FROM reviews as rev
        WHERE rev.username IN (
                            SELECT rat.username
                            FROM ratings AS rat
                            GROUP BY rat.username
                            HAVING COUNT(rat.rating_id) > 50
        )
        GROUP BY rev.username
    ) AS r
''')
```

Пользователи, которые поставили более 50 оценок, написали в среднем ~24 обзора.
